```python
# Heart Failure Prediction - Multi-model Classifier Pipeline

#  Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier

#  Load Dataset and Documentation
# Dataset Context: This dataset contains the clinical records of 299 patients who had heart failure, collected at the Ziauddin Hospital, Karachi, Pakistan, during AprilDecember 2015. The objective is to predict the likelihood of a death event.
# Column Explanation:
# age: Age of the patient (years)
# anaemia: Decrease of red blood cells or hemoglobin (boolean)
# creatinine_phosphokinase: Level of the enzyme CPK in the blood (mcg/L)
# diabetes: If the patient has diabetes (boolean)
# ejection_fraction: Percentage of blood leaving the heart at each contraction (percentage)
# high_blood_pressure: If the patient has high blood pressure (boolean)
# platelets: Platelets in the blood (kiloplatelets/mL)
# serum_creatinine: Level of serum creatinine in the blood (mg/dL)
# serum_sodium: Level of serum sodium in the blood (mEq/L)
# sex: Woman or man (binary)
# smoking: If the patient smokes (boolean)
# time: Follow-up period (days)
# DEATH_EVENT: If the patient deceased during the follow-up period (boolean)

df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
print("Dataset Shape:", df.shape)
display(df.head())

#  Exploratory Data Analysis (EDA)
print("\n--- Basic Info ---")
df.info()

print("\n--- Summary Statistics ---")
display(df.describe())

# Plot distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(x='DEATH_EVENT', data=df)
plt.title('Distribution of DEATH_EVENT')
plt.show()

# Plot distributions of numerical features
print("\n--- Distribution of Numerical Features ---")
df.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()

# Plot distributions of categorical features by target variable
print("\n--- Distribution of Categorical Features by DEATH_EVENT ---")
categorical_cols = ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking']
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, hue='DEATH_EVENT', data=df)
    plt.title(f'Distribution of {col} by DEATH_EVENT')
    plt.show()

# Plot Feature Correlation Matrix
print("\n--- Feature Correlation Heatmap ---")
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()

# Check for missing values
print("\n--- Missing Values ---")
print(df.isnull().sum())

#  Prepare Data for Modeling
# Define features (X) and target (y)
X = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']

# Split data into training and testing sets with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

# Standardize numerical features
numerical_cols = X.select_dtypes(include=np.number).columns
scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test_scaled[numerical_cols] = scaler.transform(X_test[numerical_cols])


#  Train and Evaluate Models

# Define an evaluation function to calculate and print metrics
def evaluate_model(name, y_true, y_pred, y_pred_proba):
    """Evaluates a classification model and returns key metrics."""
    print(f"\n=== {name} Evaluation ===")
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_pred_proba)

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC Score: {roc_auc:.4f}")
    print("Classification Report:")
    print(classification_report(y_true, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))

    return accuracy, precision, recall, f1, roc_auc

# List to store evaluation results
results = []

# --- Model 1: Logistic Regression ---
print("\n--- Training Logistic Regression ---")
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train_scaled, y_train)
log_preds = log_model.predict(X_test_scaled)
log_preds_proba = log_model.predict_proba(X_test_scaled)[:, 1]
results.append(
    (
        "Logistic Regression (Untuned)",
        evaluate_model(
            "Logistic Regression (Untuned)", y_test, log_preds, log_preds_proba
        ),
    )
)

# Hyperparameter Tuning for Logistic Regression
print("\n--- Tuning Logistic Regression ---")
param_dist_log = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear'],  # Use solver that supports l1 and l2
}

# assume best C = 0.1
# redifine [0.05, 0.075, 0.1, 1.25 ..]
rand_log = RandomizedSearchCV(
    LogisticRegression(max_iter=1000, random_state=42),
    param_distributions=param_dist_log,
    n_iter=10,
    cv=5,
    random_state=42,
    n_jobs=-1,
)
rand_log.fit(X_train_scaled, y_train)
best_log_model = rand_log.best_estimator_
best_log_preds = best_log_model.predict(X_test_scaled)
best_log_preds_proba = best_log_model.predict_proba(X_test_scaled)[:, 1]
results.append(
    (
        "Logistic Regression (Tuned)",
        evaluate_model(
            "Logistic Regression (Tuned)",
            y_test,
            best_log_preds,
            best_log_preds_proba,
        ),
    )
)


# --- Model 2: Naive Bayes Classifier ---
print("\n--- Training Naive Bayes ---")
nb_model = GaussianNB()
nb_model.fit(X_train_scaled, y_train)
nb_preds = nb_model.predict(X_test_scaled)
nb_preds_proba = nb_model.predict_proba(X_test_scaled)[:, 1]
results.append(("Naive Bayes", evaluate_model("Naive Bayes", y_test, nb_preds, nb_preds_proba)))


# --- Model 3: Random Forest Classifier ---
print("\n--- Training Random Forest ---")
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)  # Random forests are not sensitive to scaling
rf_preds = rf_model.predict(X_test)
rf_preds_proba = rf_model.predict_proba(X_test)[:, 1]
results.append(
    (
        "Random Forest (Untuned)",
        evaluate_model("Random Forest (Untuned)", y_test, rf_preds, rf_preds_proba),
    )
)

# Hyperparameter Tuning for Random Forest
print("\n--- Tuning Random Forest ---")
param_dist_rf = {
    'n_estimators': [50, 100, 200, 300],
    'max_depth': [3, 5, 10, 15, None],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4],
}
rand_rf = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    param_distributions=param_dist_rf,
    n_iter=10,
    cv=5,
    random_state=42,
    n_jobs=-1,
)
rand_rf.fit(X_train, y_train)
best_rf_model = rand_rf.best_estimator_
best_rf_preds = best_rf_model.predict(X_test)
best_rf_preds_proba = best_rf_model.predict_proba(X_test)[:, 1]
results.append(
    (
        "Random Forest (Tuned)",
        evaluate_model("Random Forest (Tuned)", y_test, best_rf_preds, best_rf_preds_proba),
    )
)


#  Organize and Display Evaluation Results
print("\n=== Model Evaluation Summary ===")
evaluation_summary = pd.DataFrame(results, columns=['Model', 'Metrics'])
evaluation_summary[
    ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC AUC Score']
] = pd.DataFrame(evaluation_summary['Metrics'].tolist(), index=evaluation_summary.index)
evaluation_summary = evaluation_summary.drop('Metrics', axis=1)
display(evaluation_summary.sort_values(by='ROC AUC Score', ascending=False))


#  Feature Importance Analysis

# Feature Importance for Tuned Random Forest
print("\n--- Feature Importance (Random Forest) ---")
print("Random Forest Best Parameters:", rand_rf.best_params_)
feat_importances_rf = pd.Series(
    best_rf_model.feature_importances_, index=X.columns
)
plt.figure(figsize=(10, 6))
feat_importances_rf.nlargest(10).plot(kind='barh')
plt.title("Top Feature Importances (Random Forest)")
plt.show()

```

## Feature engineering


Create new features by multiplying existing ones (e.g., 'age' and 'serum_creatinine') and apply logarithmic transformations to skewed numerical features to normalize their distributions.


```python
import pandas as pd
import numpy as np

#  Load Dataset and Documentation
# Load the heart failure dataset from the specified CSV file into a pandas DataFrame.
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')

# 1. Create a new feature by multiplying 'age' and 'serum_creatinine'
# This interaction term can capture combined effects that might be more predictive.
df['age_x_serum_creatinine'] = df['age'] * df['serum_creatinine']

# 2. Identify skewed numerical features: 'creatinine_phosphokinase' and 'serum_creatinine'
# These features have distributions that are not symmetrical, which can affect model performance.

# 3. Apply logarithmic transformations using np.log1p()
# np.log1p computes log(1 + x), which is useful for handling zero values and stabilizing variance.
df['creatinine_phosphokinase_log'] = np.log1p(df['creatinine_phosphokinase'])
df['serum_creatinine_log'] = np.log1p(df['serum_creatinine'])

# 4. Remove original skewed columns
# After creating the transformed features, the original columns are no longer needed.
df = df.drop(['creatinine_phosphokinase', 'serum_creatinine'], axis=1)

# 5. Display the first few rows of the updated DataFrame
# This helps to verify that the new features have been added correctly.
display(df.head())
```

## Advanced hyperparameter tuning


Expand the hyperparameter grids for `RandomizedSearchCV` to explore a wider range of possibilities and optimize for recall and F1-score during cross-validation.



To optimize for recall and F1-score, I will expand the hyperparameter grids for both Logistic Regression and Random Forest models and set the `scoring` parameter to 'f1' in `RandomizedSearchCV`. This will allow for a more thorough search for the best hyperparameters, specifically targeting the desired metrics. I will then fit the `RandomizedSearchCV` objects to the training data.



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Define an evaluation function to calculate and print metrics
def evaluate_model(name, y_true, y_pred, y_pred_proba):
    """Evaluates a classification model and returns key metrics."""
    print(f"\n=== {name} Evaluation ===")
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_pred_proba)

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC Score: {roc_auc:.4f}")
    print("Classification Report:")
    print(classification_report(y_true, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_true, y_pred))

    return accuracy, precision, recall, f1, roc_auc

# Prepare data
X = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

numerical_cols = X.select_dtypes(include=np.number).columns
scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test_scaled[numerical_cols] = scaler.transform(X_test[numerical_cols])

```

```python
# Expanded hyperparameter grid for Logistic Regression
# Defines a broad range of hyperparameters to test for the Logistic Regression model.
param_dist_log_expanded = {
    'C': np.logspace(-4, 4, 20),  # Regularization strength
    'penalty': ['l1', 'l2'],      # Type of regularization
    'solver': ['liblinear'],      # Algorithm to use for optimization
}

# Expanded hyperparameter grid for Random Forest
# Defines a broad range of hyperparameters to test for the Random Forest model.
param_dist_rf_expanded = {
    'n_estimators': [int(x) for x in np.linspace(start=100, stop=1000, num=10)], # Number of trees
    'max_features': ['sqrt', 'log2'], # Number of features to consider at each split
    'max_depth': [int(x) for x in np.linspace(10, 110, num=11)] + [None], # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],   # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4],     # Minimum number of samples required at each leaf node
    'bootstrap': [True, False],        # Method of sampling data points
}

# RandomizedSearchCV for Logistic Regression with F1 scoring
# This will search for the best hyperparameters for Logistic Regression, optimizing for the F1-score.
rand_log_f1 = RandomizedSearchCV(
    LogisticRegression(max_iter=1000, random_state=42),
    param_distributions=param_dist_log_expanded,
    n_iter=100,  # Number of parameter settings that are sampled
    cv=5,        # 5-fold cross-validation
    scoring='f1',# Optimize for F1-score
    random_state=42,
    n_jobs=-1,   # Use all available CPU cores
)
# Fit the RandomizedSearchCV object to the scaled training data.
rand_log_f1.fit(X_train_scaled, y_train)
# Get the best model found by the search.
best_log_model_f1 = rand_log_f1.best_estimator_
# Make predictions on the test set using the best model.
best_log_preds_f1 = best_log_model_f1.predict(X_test_scaled)
# Get the probability predictions for the positive class.
best_log_preds_proba_f1 = best_log_model_f1.predict_proba(X_test_scaled)[:, 1]

# RandomizedSearchCV for Random Forest with F1 scoring
# This will search for the best hyperparameters for Random Forest, optimizing for the F1-score.
rand_rf_f1 = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    param_distributions=param_dist_rf_expanded,
    n_iter=100,
    cv=5,
    scoring='f1',
    random_state=42,
    n_jobs=-1,
)
# Fit the RandomizedSearchCV object to the training data.
rand_rf_f1.fit(X_train, y_train)
# Get the best model found by the search.
best_rf_model_f1 = rand_rf_f1.best_estimator_
# Make predictions on the test set using the best model.
best_rf_preds_f1 = best_rf_model_f1.predict(X_test)
# Get the probability predictions for the positive class.
best_rf_preds_proba_f1 = best_rf_model_f1.predict_proba(X_test)[:, 1]

# Evaluate the new tuned models
# Print the evaluation metrics for the tuned Logistic Regression model.
print("\n--- Logistic Regression (Tuned for F1) ---")
evaluate_model("Logistic Regression (Tuned for F1)", y_test, best_log_preds_f1, best_log_preds_proba_f1)

# Print the evaluation metrics for the tuned Random Forest model.
print("\n--- Random Forest (Tuned for F1) ---")
evaluate_model("Random Forest (Tuned for F1)", y_test, best_rf_preds_f1, best_rf_preds_proba_f1)
```

**age_x_serum_creatinine**

**Description**: This feature is the product of age and serum_creatinine.

*Rationale*: The interaction between age and serum creatinine levels can be a more powerful predictor of heart failure risk than either feature alone. As people age, their kidney function may decline, leading to higher serum creatinine levels. This combined effect could be a significant indicator of health decline.

**creatinine_phosphokinase_log**

Description: This is the log-transformed value of creatinine_phosphokinase.

*Rationale*: The creatinine_phosphokinase feature has a skewed distribution with a long tail of high values. Applying a logarithmic transformation helps to normalize this distribution, making it more suitable for linear models and reducing the influence of outliers.

**serum_creatinine_log**

Description: This is the log-transformed value of serum_creatinine.
Rationale: Similar to creatinine_phosphokinase, the serum_creatinine feature is also skewed. A logarithmic transformation helps to stabilize the variance and make the distribution more symmetric, which can improve the performance of many machine learning models.
